```python
def solve():
    n = int(input())  #   
    a = list(map(int, input().split()))  #    

    a.sort()  #     

    max_diff_sum = 0  #       

    for i in range(n // 2):  #     ( )
        max_diff_sum += a[n - 1 - i] - a[i]  #         

    print(max_diff_sum)  #    

solve()
```

```python
def solve():
    n = int(input())
    a = list(map(int, input().split()))
    a.sort()

    res = 0
    for i in range(n // 2):
        res += abs(a[i] - a[n - i - 1])
    print(res)

solve()
```
