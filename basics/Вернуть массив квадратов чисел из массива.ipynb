# Дан массив целых чисел, отсортированный по возрастанию. Необходимо вернуть массив квадратов этих чисел, отсортированный так же по возрастанию.

#### Сложность O(n)

```python
def отсортированные_квадраты(массив):

  n = len(массив)
  результат = [0] * n
  левый = 0
  правый = n - 1
  индекс = n - 1

  while левый <= правый:
    if abs(массив[левый]) > abs(массив[правый]):
      результат[индекс] = массив[левый] ** 2
      левый += 1
    else:
      результат[индекс] = массив[правый] ** 2
      правый -= 1
    индекс -= 1

  return результат
```

```python
# Пример использования:
массив = [-4, -1, 0, 3, 10]
квадраты = отсортированные_квадраты(массив)
print(квадраты)  # Вывод: [0, 1, 9, 16, 100]
```
