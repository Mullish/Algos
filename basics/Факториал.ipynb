# Факториал числа. Напишите рекурсивную функцию, которая вычисляет факториал заданного числа.

```python
def factorial(n):
    if n < 0:
        raise ValueError("Факториал не определен для отрицательных чисел")
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
# Пример использования
print(factorial(5))  # Вывод: 120
```
