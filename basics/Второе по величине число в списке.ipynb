# Напишите функцию, которая находит второе по величине число в списке.

```python
def second_largest(numbers):
    unique_numbers = list(set(numbers))
    unique_numbers.sort()
    return unique_numbers[-2] if len(unique_numbers) >= 2 else None
```

```python
# Пример использования
numbers = [5, 1, 2, 3, 5]
print(second_largest(numbers))  # Вывод: 3
```
