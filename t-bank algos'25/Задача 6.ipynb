```python
def solve():
    n = int(input())  # Читаем количество сотрудников
    a = list(map(int, input().split()))  # Читаем массив ростов сотрудников

    a.sort()  # Сортируем массив ростов по возрастанию

    max_diff_sum = 0  # Инициализируем переменную для хранения максимальной суммы разностей

    for i in range(n // 2):  # Итерируемся по половине массива (целочисленное деление)
        max_diff_sum += a[n - 1 - i] - a[i]  # Добавляем разницу между самым высоким и самым низким сотрудником

    print(max_diff_sum)  # Выводим максимальную сумму разностей

solve()
```

```python
def solve():
    n = int(input())
    a = list(map(int, input().split()))
    a.sort()

    res = 0
    for i in range(n // 2):
        res += abs(a[i] - a[n - i - 1])
    print(res)

solve()
```
