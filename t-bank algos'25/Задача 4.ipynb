```python
def solve():
    n = int(input())
    arr = list(map(int, input().split()))

    count = 0

    for i in range(n - 2):
        for j in range(i + 2, n):
            sub_array = arr[i:j+1]

            if len(sub_array) < 3:
                continue

            found = False
            for x in range(len(sub_array)):
                for y in range(len(sub_array)):
                    for z in range(len(sub_array)):
                        if x != y and x != z and y != z:
                            a = sub_array[x]
                            b = sub_array[y]
                            c = sub_array[z]

                            if b - a == c - b:
                                found = True
                                break
                    if found:
                        break
                if found:
                    break

            if found:
                count += 1

    print(count)

solve()
```
