```python
def min_cost_to_balance(n, a, b, s):
    balance = 0
    min_balance = 0
    for ch in s:
        if ch == '(':
            balance += 1
        else:
            balance -= 1
        min_balance = min(min_balance, balance)

    need_open = -min_balance
    need_close = balance

    swaps = min(need_open, need_close)
    left = need_open - swaps
    right = need_close - swaps

    cost_swap = a
    cost_replace = b

    if cost_swap > 2 * cost_replace:
        return (need_open + need_close) * cost_replace
    else:
        return swaps * cost_swap + (left + right) * cost_replace

n, a, b = map(int, input().split())
s = input().strip()

print(min_cost_to_balance(n, a, b, s))
```
