# Напишите функцию, которая находит первый уникальный символ в строке.

```python
def first_unique_char(s):
    char_count = {}
    for char in s:
        char_count[char] = char_count.get(char, 0) + 1
    for char in s:
        if char_count[char] == 1:
            return char
    return None
```

- char_count.get(char, 0) пытается получить текущее количество вхождений символа char. Если этот символ раньше не встречался, возвращается значение 0.
   - Затем к текущему значению добавляется 1, чтобы учесть текущее вхождение символа. Результат сохраняется обратно в char_count с ключом char.

```python
# Пример использования
string = "leetcode"
print(first_unique_char(string))  # Вывод: l
```
