```python
def main():
    n = int(input())

    starts = []
    intervals = []
    for _ in range(n):
        start, interval = map(int, input().split())
        starts.append(start)
        intervals.append(interval)

    q = int(input())

    numbers = []
    for _ in range(q):
        branch, time = map(int, input().split())
        branch -= 1

        start = starts[branch]
        interval = intervals[branch]

        if time <= start:
            a = start
        else:
            diff = time - start
            num_intervals = (diff + interval - 1) // interval
            next_time = start + num_intervals * interval
            a = next_time

        numbers.append(a)

    for value in numbers:
        print(value)

if __name__ == "__main__":
    main()
```
