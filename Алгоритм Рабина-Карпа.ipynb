# Поиск индекса первого вхождения подстроки в строку (алгоритм Рабина-Карпа).



```python
def rabin_karp(text, pattern):
    n, m = len(text), len(pattern)
    if m > n:
        return -1

    base = 256
    mod = 101
    hash_pattern = 0
    hash_text = 0
    h = 1

    for i in range(m-1):
        h = (h * base) % mod

    for i in range(m):
        hash_pattern = (base * hash_pattern + ord(pattern[i])) % mod
        hash_text = (base * hash_text + ord(text[i])) % mod

    for i in range(n - m + 1):
        if hash_pattern == hash_text:
            if text[i:i+m] == pattern:
                return i
        if i < n - m:
            hash_text = (base * (hash_text - ord(text[i]) * h) + ord(text[i + m])) % mod
            if hash_text < 0:
                hash_text += mod
    return -1
```

```python
print(rabin_karp("abracadabra", "cad"))  # 4
```
