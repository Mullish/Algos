# Поиск дубликатов. Дан список целых чисел nums. Найти все повторяющиеся элементы списка и вернуть их в виде множества.

```python
def find_duplicates(nums):
    seen = set()
    duplicates = set()

    for num in nums:
        if num in seen:
            duplicates.add(num)
        else:
            seen.add(num)

    return duplicates
```

```python
# Тестирование
print(find_duplicates([1, 2, 3, 4, 2, 5, 6, 7, 8, 1]))  # {1, 2}
```
