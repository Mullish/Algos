# Поиск самого длинного общего префикса среди нескольких строк. Для заданного списка строк верните самый длинный общий префикс всех строк. Если общих префиксов нет — вернуть пустую строку.

```python
def longest_common_prefix(strs):
    if not strs:
        return ""

    prefix = strs[0]

    for i in range(len(prefix)):
        char = prefix[i]

        for other_str in strs[1:]:
            if i >= len(other_str) or other_str[i] != char:
                return prefix[:i]

    return prefix
```

```python
# Тестирование
print(longest_common_prefix(["flower","flow","flight"]))  # 'fl'
```
